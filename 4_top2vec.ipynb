Getting the tweets processed.

```python
import pandas as pd

tweets = pd.read_json("/content/drive/MyDrive/UCSP/Data-Science-Topics/Data/tweets_text_processed.json", orient="split")
tweets
```

```python
tweets_by_month = tweets.copy()
tweets_by_month["month"] = tweets_by_month.date.dt.month
tweets_by_month
```

```python
tweets_list = []

for i, _tweets in tweets_by_month.groupby('month'):
  tweets_list.append(pd.DataFrame(_tweets[['date', 'text_processed']]))
```

```python
temp = tweets_list[3].reset_index(drop=True)  # Abril
temp
```

Generate two separate lists of tweets and dates.

```python
_tweets = []
_dates = []

for i in range(len(temp)):
  _tweets.append(temp['text_processed'][i])
  _dates.append(temp['date'][i])

print(len(_tweets))
print(len(_dates))
```

Run the Top2Vec

```python
from top2vec import Top2Vec

model = Top2Vec(_tweets, embedding_model='universal-sentence-encoder-multilingual')
model.save("/content/drive/MyDrive/UCSP/Data-Science-Topics/Final-project/Top2Vec-models/model-09")
```
